```python
from IPython.display import Image

from selenium import webdriver
from selenium.webdriver.firefox.options import Options
options = Options()
options.add_argument('-headless')
firefox_profile = webdriver.FirefoxProfile()
firefox_profile.set_preference("permissions.default.stylesheet", 2)  #  1
firefox_profile.set_preference("permissions.default.image", 2)  # 
firefox_profile.set_preference("dom.ipc.plugins.enabled.libflashplayer.so", 2)  # flash
firefox_profile.update_preferences()  # 

driver = webdriver.Firefox(firefox_profile, timeout=10, firefox_options=options)
driver.set_window_size(1024, 5555)

for i in range(100):
    #_____________________url___________________________
    pn=i*10
    if i==0:
        pn=1
    usm='100'
    word=''
    sa='np'
    rsv_pq=''
    rsv_t=''
    rqid=''
    #________________________________________________
    url = 'https://m.baidu.com/s?pn='+pn+'&usm='+usm+'&word='+word+'&sa='+sa+'&rsv_pq='+rsv_pq+'&rsv_t='+rsv_t+'&rqid='+rqid
    driver.get (url)
    contant = driver.page_source
    driver.save_screenshot('csdn2.png')
    driver.delete_all_cookies()#cookie
    #driver.delete_cookie('my_cookie')#cookie
    driver.close()
Image("csdn2.png")
```
