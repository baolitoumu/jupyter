```python
from IPython.display import Image

from selenium import webdriver
from selenium.webdriver.firefox.options import Options
options = Options()
options.add_argument('-headless')
firefox_profile = webdriver.FirefoxProfile()





ip = "177.69.195.4"

duankou = int(3128)#这里记得类型转换，这坑踩了2天
firefox_profile.set_preference("network.proxy.type", 1)
# ip及其端口号配置为 http 协议代理
firefox_profile.set_preference("network.proxy.http", ip)
firefox_profile.set_preference("network.proxy.http_port", duankou)

# 所有协议共用一种 ip 及端口，如果单独配置，不必设置该项，因为其默认为 False
firefox_profile.set_preference("network.proxy.share_proxy_settings", True)

driver = webdriver.Firefox(firefox_profile, timeout=10, firefox_options=options)
driver.set_window_size(1024, 5555)

```

```python
for i in range(1):
    #_____________________url参数___________________________
    pn='999'
    usm='100'
    word='免费代理'
    sa='np'
    rsv_pq=''
    rsv_t=''
    rqid=''
    #_____________________启动浏览器___________________________
    url = 'https://m.baidu.com/s?pn='+pn+'&usm='+usm+'&word='+word+'&sa='+sa+'&rsv_pq='+rsv_pq+'&rsv_t='+rsv_t+'&rqid='+rqid
    driver.get (url)
    contant = driver.page_source
    driver.save_screenshot('csdn2.png')
    driver.delete_all_cookies()#删除当前会话的所有cookie
    #driver.delete_cookie('my_cookie')#删除一个指定的cookie
    driver.close()
Image("csdn2.png")
```

```python
from selenium import webdriver
from selenium.webdriver.firefox.firefox_profile import FirefoxProfile

profile = FirefoxProfile()
# 激活手动代理配置（对应着在 profile（配置文件）中设置首选项）
profile.set_preference("network.proxy.type", 1)
# ip及其端口号配置为 http 协议代理
profile.set_preference("network.proxy.http", "118.190.95.35")
profile.set_preference("network.proxy.http_port", 9001)

# 所有协议共用一种 ip 及端口，如果单独配置，不必设置该项，因为其默认为 False
profile.set_preference("network.proxy.share_proxy_settings", True)

# 默认本地地址（localhost）不使用代理，如果有些域名在访问时不想使用代理可以使用类似下面的参数设置
# profile.set_preference("network.proxy.no_proxies_on", "localhost")

# 以代理方式启动 firefox
firefox  = webdriver.Firefox(profile)
firefox.get('https://ip.cn/')
firefox.save_screenshot('csdn2.png')
firefox.delete_all_cookies()#删除当前会话的所有cookie
firefox.close()
Image("csdn2.png")
```
