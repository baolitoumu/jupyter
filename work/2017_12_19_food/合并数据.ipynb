```python
# The usual preamble
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
import time
# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('max_colwidth',200)
pd.set_option('display.max_columns', 90)
pd.set_option('display.max_rows', 90)
writer = pd.ExcelWriter('合并数据.xlsx')
```

```python
xls_file = pd.ExcelFile('20171219food.xlsx')
```

```python
sheetlist = xls_file.sheet_names 
print(sheetlist)
len(sheetlist)
#sheetlist.remove('分类商品数量')
#print(sheetlist)
```

```python
time_now=datetime.now()
x=[]
for i in range(len(sheetlist)):
    tn = datetime.now()
    
    
    locals()['d'+str(i)]=i
    print ('d'+str(i))
    print(sheetlist[i])
    x.append('d'+str(i))
    x[i] = pd.read_excel('20171219food.xlsx',i)
    x[i] = x[i].replace(0,np.nan)    
    x[i] = x[i].fillna(0)
    to = datetime.now()
    tx = ((to-tn).seconds)/60
    print('当前耗时++++',tx,'分钟')

time_out=datetime.now()
time_total = ((time_out-time_now).seconds)/60
print('总耗时++++',time_total,'分钟')
```

```python

adn = x[0]
for i in range((len(sheetlist))-1):
    adn = pd.merge(adn,x[i+1], how='left', on='商户id')  
    
    

#adn = pd.concat(x)
#adn = pd.merge(x, how='left', on='商户id') 
adn = adn.replace(0,np.nan)    
adn = adn.fillna(0)
#print(adn)
print(adn)
```

```python
adn.to_excel(writer,'合并后数据')
```

```python
writer.save()
```
