```python
from torch.autograd import Variable
import torch
from torch import nn
import pandas as pd
import matplotlib.pyplot as plt
import torch.nn.functional as F
import numpy as np
import random
```

```python
price_data = pd.read_csv(open('dataset.csv'))
train_data = torch.from_numpy(price_data.iloc[1:151,2:3].values)#取第1-148行，2-10列
#test_data = torch.from_numpy((price_data.iloc[100:198,2:4].values).T) #取第149-199行，2-10列
test_data = torch.from_numpy((price_data.iloc[151:302,2:3].values))
```

```python
train_data = Variable(train_data, requires_grad=True).type(torch.FloatTensor)
test_data = Variable(test_data).type(torch.FloatTensor)
print('train_data_size:',train_data.size())
print('test_data_size:',test_data.size())
```

```python
class Net(nn.Module):
    def __init__(self,D_in,H,D_out):
        super(Net, self).__init__()  
        self.linear1 =nn.Linear(D_in,H)
        self.linear2 = nn.Linear(H,D_out)
        
    def forward(self,train_data):
        train_data = F.relu(self.linear1(train_data))
        train_data = self.linear2(train_data)
        
        return train_data
module = Net(D_in=1, H=2, D_out=1)
```

```python
print('第一层权重',module.linear1.weight)# 
print('第一层biases',module.linear1.bias)#b
print('第二层权重',module.linear2.weight)# w
print('第二层biases',module.linear2.bias)#b
#print(module.gradWeight)
#print(module.gradBias)
```

```python
loss_fn = torch.nn.MSELoss(size_average=False)
optimizer = torch.optim.Adam(module.parameters(), lr=1) 
loopnum = 222
```

```python
for t in range(loopnum):
    a = random.randint(0,148)
    trian = train_data[a:a+1]
    test = test_data[a:a+1]
    
    ########################
    pre_data = module(trian)
    loss = loss_fn(pre_data,test)
    
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    print('训练值',trian)
    print('目标值',test)
    print('预测值',pre_data)
    print('损失值',loss)
    print('当前第##########',t,'###############次训练')
    
    
    
```
